```python
import numpy as np
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup fertig")
```

```python
#Quelle: https://www.rki.de/DE/Content/InfAZ/N/Neuartiges_Coronavirus/Daten/Altersverteilung.html;jsessionid=EC085F189E46636FFABE01920AF9C225.internet072?nn=13490888
#Stand: 22.06.2021
```

```python
age_filepath="Altersverteilung.xlsx"
age_data=pd.read_excel(age_filepath,parse_dates=True)  #Alterspezifische 7-Tage-Inzidenz nach Meldewoche
age_data.head()
```

```python
age1=age_data.iloc[1:]
plt.figure(figsize=(10,2))
sns.catplot(data=age1, kind="bar", x="2020_20",y="Altersgruppe")
sns.catplot(data=age1, kind="bar", x="2021_24", y="Altersgruppe")
```

```python
#Vermutung: Die Inzidenz hat sich im Laufe des Jahres verlagert:
#Waren am Anfang noch viele im hohen Alter betroffen, so sind in KW 20 in 2021 (Montag, 17. â€“ Sonntag, 23. Mai 2021)
#viele junge Menschen betroffen.
```

```python
fig, axes = plt.subplots(2, 4, sharex=True, figsize=(20,8))
fig.suptitle('Vergleich der 7-Tages-Inzidenz 2020 mit 2021 exemplarisch (von 90+ Jahre links, bis 0-4 Jahre rechts)')
age1=age_data.iloc[1:] #Gesamtzeile wegschneiden

#2020
#KW 13 2020
sns.barplot(ax=axes[0,0],data=age1, x="Altersgruppe", y="2020_13")
axes[0,0].set(xlabel="Altersgruppen",ylabel="Inzidenz",ylim=[0,250],title="KW 13 2020")
#KW 14 2020
sns.barplot(ax=axes[0,1], x=age1.index, y=age1['2020_14'])
axes[0,1].set(xlabel="Altersgruppen",ylabel="Inzidenz",ylim=[0,250],title="KW 14 2020")
#KW 15 2020
sns.barplot(ax=axes[0,2], x=age1.index, y=age1['2020_15'])
axes[0,2].set(xlabel="Altersgruppen",ylabel="Inzidenz",ylim=[0,250],title="KW 15 2020")
#KW 16 2020
sns.barplot(ax=axes[0,3], x=age1.index, y=age1['2020_16'])
axes[0,3].set(xlabel="Altersgruppen",ylabel="Inzidenz",ylim=[0,250],title="KW 16 2020")

#2021
#KW 13 2021
sns.barplot(ax=axes[1,0], x=age1.index, y=age1['2021_13'])
axes[1,0].set(xlabel="Altersgruppen",ylabel="Inzidenz",ylim=[0,250],title="KW 13 2021")
#KW 14 2021
sns.barplot(ax=axes[1,1], x=age1.index, y=age1['2021_14'])
axes[1,1].set(xlabel="Altersgruppen",ylabel="Inzidenz",ylim=[0,250],title="KW 14 2021")
#KW 15 2021
sns.barplot(ax=axes[1,2], x=age1.index, y=age1['2021_15'])
axes[1,2].set(xlabel="Altersgruppen",ylabel="Inzidenz",ylim=[0,250],title="KW 15 2021")
#KW 16 2021
sns.barplot(ax=axes[1,3], x=age1.index, y=age1['2021_16'])
axes[1,3].set(xlabel="Altersgruppen",ylabel="Inzidenz",ylim=[0,250],title="KW 16 2021")
```
